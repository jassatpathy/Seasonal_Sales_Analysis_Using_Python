```python
# Exploratory Data Analysis
```

```python
# Gender
```

```python
# plotting a bar chart for Gender and it's count
ax = sns.countplot(
    x='Gender',
    data=df,
    palette=['red', 'blue']  # Assign colors in order
)

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
# # plotting a bar chart for gender vs total amount

# Group and sum

sales_gen = (
    df.groupby(['Gender'], as_index=False)['Amount']
      .sum()
      .sort_values(by='Amount', ascending=False)
)

# Create bar plot with custom colors
ax = sns.barplot(
    x='Gender',
    y='Amount',
    data=sales_gen,
    palette=['blue', 'red']  # custom colors
)

# Add value labels above bars
for p in ax.patches:
    ax.annotate(
        f'{p.get_height():,.0f}',       # format with commas
        (p.get_x() + p.get_width() / 2, p.get_height()),  
        ha='center', va='bottom',
        fontsize=10, fontweight='bold',
        color='black'
    )

plt.title("Total Amount by Gender")
plt.ylabel("Total Amount")
plt.xlabel("Gender")
plt.show()
```

```python
# From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men
```

```python
# Age
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Match palette keys to actual Gender values in df
palette = {'M': 'blue', 'F': 'red'}

ax = sns.countplot(
    data=df,
    x='Age Group',
    hue='Gender',
    palette=palette
)

# Add labels on top of bars
for bars in ax.containers:
    ax.bar_label(bars)

plt.title("Age Group vs Gender")
plt.show()
```

```python
# Total Amount vs Age Group

# Group and sum
sales_age = (
    df.groupby(['Age Group'], as_index=False)['Amount']
      .sum()
      .sort_values(by='Amount', ascending=False)
)

# Define custom colors
colors = ['green', 'red', 'blue', 'yellow', 'orange', 'navy', 'pink']

# Create bar plot
ax = sns.barplot(
    x='Age Group',
    y='Amount',
    data=sales_age,
    palette=colors
)

# Add value labels above bars
for p in ax.patches:
    ax.annotate(
        f'{p.get_height():,.0f}',  # comma formatted numbers
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center', va='bottom',
        fontsize=10, fontweight='bold'
    )

plt.title("Total Amount by Age Group")
plt.ylabel("Total Amount")
plt.xlabel("Age Group")
plt.show()
```

```python
# From above graphs we can see that most of the buyers are of age group between 26-35 yrs female
```

```python
# State
```

```python
# Top 10 states by total orders

sales_state = (
    df.groupby(['State'], as_index=False)['Orders']
      .sum()
      .sort_values(by='Orders', ascending=False)
      .head(10)
)

# Set figure size
sns.set(rc={'figure.figsize': (15, 5)})

# Define 10 distinct, visually pleasing colors
colors = [
    'green', 'red', 'blue', 'orange', 'purple',
    'cyan', 'magenta', 'gold', 'navy', 'pink'
]

# Create bar plot
ax = sns.barplot(
    data=sales_state,
    x='State',
    y='Orders',
    palette=colors
)

# Add labels above bars
for p in ax.patches:
    ax.annotate(
        f'{int(p.get_height()):,}',   # formatted with commas
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center', va='bottom',
        fontsize=10, fontweight='bold'
    )

plt.title("Total Orders from Top 10 States")
plt.xlabel("State")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# total amount/sales from top 10 states
#  From this graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

sales_state = (
    df.groupby(['State'], as_index=False)['Amount']
      .sum()
      .sort_values(by='Amount', ascending=False)
      .head(10)
)

# Set figure size
sns.set(rc={'figure.figsize': (15, 5)})

# Define 10 distinct colors
colors = [
    'green', 'red', 'blue', 'orange', 'purple',
    'cyan', 'magenta', 'gold', 'navy', 'pink'
]

# Create bar plot
ax = sns.barplot(
    data=sales_state,
    x='State',
    y='Amount',
    palette=colors
)

# Add value labels on top of bars
for p in ax.patches:
    ax.annotate(
        f'{p.get_height():,.0f}',   # comma-formatted values
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center', va='bottom',
        fontsize=10, fontweight='bold'
    )

plt.title("Total Amount from Top 10 States")
plt.xlabel("State")
plt.ylabel("Total Amount")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Marital Status
```

```python
sns.set(rc={'figure.figsize': (7, 5)})

# Two colors for the two categories
colors = ['green', 'orange']

ax = sns.countplot(
    data=df,
    x='Marital_Status',
    palette=colors
)

# Add labels above each bar
for bars in ax.containers:
    ax.bar_label(bars)

plt.title("Count of Marital Status")
plt.xlabel("Marital Status")
plt.ylabel("Count")
plt.show()
```

```python

# Group data
sales_state = (
    df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount']
      .sum()
      .sort_values(by='Amount', ascending=False)
)

sns.set(rc={'figure.figsize': (6, 5)})

# Map gender to colors
# Adjust keys depending on whether Gender is 'Male'/'Female' or 'M'/'F'
palette = {'M': 'green', 'F': 'orange'}  # or {'Male': 'green', 'Female': 'orange'}

ax = sns.barplot(
    data=sales_state,
    x='Marital_Status',
    y='Amount',
    hue='Gender',
    palette=palette
)

# Add labels above each bar
for bars in ax.containers:
    ax.bar_label(bars)

plt.title("Total Amount by Marital Status and Gender")
plt.ylabel("Total Amount")
plt.xlabel("Marital Status")
plt.show()
```

```python
# From above graphs we can see that most of the buyers are married (women) and they have high purchasing power
```

```python
# Product Category
```

```python
sns.set(rc={'figure.figsize': (20, 5)})
ax = sns.countplot(data=df, x='Product_Category', palette='tab10')

for bars in ax.containers:
    ax.bar_label(bars)

plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# From this graphs we can see that most of the sold products are from Food, Clothing and Electronics category
sales_product = (
    df.groupby(['Product_Category'], as_index=False)['Amount']
      .sum()
      .sort_values(by='Amount', ascending=False)
      .head(10)
)

sns.set(rc={'figure.figsize': (20, 5)})
sns.barplot(
    data=sales_product,
    x='Product_Category',
    y='Amount',
    palette='tab10'
)

plt.xticks(rotation=45, ha='right')
plt.show()

```

```python
# Top selling product id

sales_state = (
    df.groupby(['Product_ID'], as_index=False)['Orders']
      .sum()
      .sort_values(by='Orders', ascending=False)
      .head(10)
)

sns.set(rc={'figure.figsize': (20, 5)})

ax = sns.barplot(
    data=sales_state,
    x='Product_ID',
    y='Orders',
    palette='tab10'  # 10 distinct colors
)

# Add labels above bars
for p in ax.patches:
    ax.annotate(
        f'{int(p.get_height())}',
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center', va='bottom',
        fontsize=10, fontweight='bold'
    )

plt.xticks(rotation=45, ha='right')
plt.title("Top 10 Products by Orders")
plt.show()
```

```python


fig1, ax1 = plt.subplots(figsize=(12, 7))

# Data for top 10 products by orders
top_products = (
    df.groupby('Product_ID')['Orders']
      .sum()
      .nlargest(10)
      .sort_values(ascending=False)
)

# Choose 10 distinct colors
colors = ['green', 'red', 'blue', 'orange', 'purple',
          'cyan', 'magenta', 'gold', 'navy', 'pink']

# Plot with colors
top_products.plot(kind='bar', color=colors, ax=ax1)

# Add value labels on top of bars
for i, v in enumerate(top_products):
    ax1.text(i, v, f'{v}', ha='center', va='bottom', fontsize=10, fontweight='bold')

ax1.set_title("Top 10 Products by Orders")
ax1.set_xlabel("Product ID")
ax1.set_ylabel("Orders")
plt.xticks(rotation=45, ha='right')
plt.show()
```
